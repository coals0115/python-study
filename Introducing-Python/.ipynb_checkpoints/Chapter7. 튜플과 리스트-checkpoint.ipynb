## 여러 시퀀스 순회하기: zip()

```python
days = ['Monday','Tuesday','Wednesday']
fruits = ['banana','orange','peach']
drinks = ['coffee','tea','beer']
desserts=['tiramisu','ice cream','pie','pudding']
for day,fruit,drink,dessert in zip(days, fruits, drinks,desserts):
    print(day, "drink",drink, "-eat",fruit, "- enjoy", dessert)

# Monday drink coffee -eat banana - enjoy tiramisu
# Tuesday drink tea -eat orange - enjoy ice cream
# Wednesday drink beer -eat peach - enjoy pie
```

+) 여러 시퀀스중 가장 짧은 시퀀스가 완료되면 zip()은 멈춘다**.**

```python
english = 'Monday', 'Tuesday', 'Wednesday'
french = 'Lundi', 'Mardi', 'Mercredi'

list(zip(english, french)
# [('Monday', 'Lundi'), ('Tuesday', 'Mardi'), ('Wednesday', 'Mercredi')]

dict(zip(english, french))
# {'Monday': 'Lundi', 'Tuesday': 'Mardi', 'Wednesday': 'Mercredi'}
```

## 리스트 컴프리헨션

리스트 값을 생성하는 첫 번째 number 변수 필요. 이것은 순회 결과를 number_list 변수에 넣어둔다.

두 번째 number 변수는 for문의 일부다. 

```python
**[표현식 for 항목 in 순회 가능한 객체]**
number_list = [number for number in range(1, 6)] # [1, 2, 3, 4, 5]
number_list = [number-1 for number in range(1, 6)] # [0, 1, 2, 3, 4] 
```

```python
**[표현식 for 항목 in 순회 가능한 객체 if 조건]**
a_list = [number for number in range(1, 6) if number % 2 == 1] # [1, 3, 5]

# 컴프리헨션이 더 단순
a_list = []
for number in range(1, 6):
	if number % 2 == 1:
		a_list.append(number)
```

---

루프가 중첩될 수 있는 것처럼 컴프리헨션에서 루프에 상응하는 하나 이상의 for ... 절을 사용할 수 있다.

```python
# 일반적인 중첩 루프 결과
rows = range(1,4)
cols = range(1,3)
for row in rows:
    for col in cols:
        print(row, col)

# 컴프리헨션 사용 (row,col) 튜플 리스트를 만들어서 cells 변수에 할당.
rows = range(1,4)
cols = range(1,3)
cells = [(row,col) for row in rows for col in cols]
for cell in cells:
    print(cell)

# 그리고 cells 리스트를 순회한 것처럼，
# 각 튜플로부터 row와 col의 값만 출력하기 위해 튜플 **언패킹**을 할 수 있다.
for row, col in cells:
    print(row, col)
```

## 튜플 vs 리스트

튜플 = 불변, append(), insert()같은 함수 없음, 수정할 수 없기 때문에 함수의 수가 매우 적음

리스트 = 가변

**<튜플을 사용하는 이유>**

- 튜플은 더 적은 공간을 사용한다.
- 실수로 튜플의 항목이 손상될 염려가 없다
- 튜플을 딕셔너리 키로 사용할 수 있다(딕셔너리는 8장 참조).
- 네임드튜플(10.11 절 참조)은 객체의 단순한 대안이 될 수 있다.

일반적으로는 리스트(List)와 딕셔너리(Map)를 더 많이 사용한다.

## 튜플 컴프리헨션은 없다.

List, Set, Map에는 컴프리헨션이 있으나 문자열과 튜플과 같은 불변은 다른 방법으로 만들어져야 한다.

()안에 있는 것은 `**제너레이터 컴프리헨션**`임. 제너레이터 객체를 반환한다.

```python
number_thing = (number for number in range(1,  6)) # <class 'generator'>
```
